```python
class Calculator:
    def __init__(self, n): #constructor
        self.n = n  #instance variable
    
    def square(self):#method
        print(f"The square is {self.n*self.n}")#

    def cube(self):#
        print(f"The cube is {self.n*self.n*self.n}") 

    def squareroot(self):
        print(f"The squareroot is {self.n**1/2}")

a = Calculator(4)
a.square()
a.cube()
a.squareroot()
```
