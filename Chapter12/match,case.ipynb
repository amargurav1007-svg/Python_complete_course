```python
def http_status(status): 
    match status:  # Python 3.10+ only  
        case 200: 
            return "OK"  # HTTP Status 200: OK      
        case 404: 
            return "Not Found" 
        case 500: 
            return "Internal Server Error" 
        case _: 
            return "Unknown status"  


print(http_status(500))
```
